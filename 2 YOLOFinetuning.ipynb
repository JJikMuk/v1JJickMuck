```python
# 1) YOLO 모델 파인튜닝 훈련 및 자동 저장

import os
from ultralytics import YOLO

# ===============================
# 사용자 입력: 기존 모델 경로
# ===============================
pretrained_model_epoch = 0
pretrained_model_path = f"Model/YOLO_nutrition_detection{pretrained_model_epoch}/weights/best.pt"
default_model = "yolov8n.pt"

epochs = 200

# ===============================
# 모델 불러오기
# ===============================
if os.path.exists(pretrained_model_path):
    print(f"기존 모델 발견, 불러와서 이어서 학습합니다: {pretrained_model_path}")
    model = YOLO(pretrained_model_path)
else:
    print(f"기존 모델 없음, 사전학습 모델로 새로 학습합니다: {default_model}")
    model = YOLO(default_model)

# ===============================
# 학습
# ===============================
model.train(
    data="data.yaml",                 # YOLO 데이터 yaml
    epochs=epochs - pretrained_model_epoch,                    # 학습 반복 횟수
    imgsz=640,                        # 입력 이미지 크기
    batch=8,                          # 배치 사이즈
    project="Model/",                 # 학습 결과 저장 폴더
    name=f"YOLO_nutrition_detection{epochs}",  # 학습 이름
    exist_ok=True                     # 같은 이름 폴더가 있으면 덮어쓰기
)
```
